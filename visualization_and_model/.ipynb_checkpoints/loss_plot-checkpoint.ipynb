```python
import pickle
import torch
NAME_LIST = []
RECORD_LIST = [
    'domain_specific/loss_Botnet_1.txt',
    'domain_specific/loss_Dropper_1.txt',
    'domain_specific/loss_Ransomware_1.txt',
    'domain_specific/loss_Backdoor_1.txt',
    'domain_specific/loss_Spyware_1.txt',
    'domain_specific/loss_Virus_1.txt',
    'domain_specific/loss_Adware_1.txt',
    'domain_specific/loss_Misc_1.txt',
    'domain_specific/loss_Rootkit_1.txt'
    
]
POINTER_LIST = [

    'domain_specific/loss_Botnet_pointer_1.txt',
    'domain_specific/loss_Dropper_pointer_1.txt',
    'domain_specific/loss_Ransomware_pointer_1.txt',
    'domain_specific/loss_Backdoor_pointer_1.txt',
    'domain_specific/loss_Spyware_pointer_1.txt',
    'domain_specific/loss_Virus_pointer_1.txt',
    'domain_specific/loss_Adware_pointer_1.txt',
    'domain_specific/loss_Misc_pointer_1.txt',
    'domain_specific/loss_Rootkit_pointer_1.txt'
]
name_list = ['Botnet','Dropper','Ransomware','Backdoor','Spyware','Virus','Adware','Misc','Rootkit']
DATA_ALL = []
DATA_POINTE_ALL = []
for i in range(0, len(POINTER_LIST)):
    print(i)
    loss_data = []
    loss_data2 = []
    with open(RECORD_LIST[i], 'rb') as filehandle:
        loss_list = pickle.load(filehandle)
    with open(POINTER_LIST[i], 'rb') as filehandle:
        loss_list2 = pickle.load(filehandle)
        
    for j in range(len(loss_list)):
        try:
            loss_data.append(loss_list[j].item())
            loss_data2.append(loss_list2[j])
        except:
            continue
    DATA_ALL.append(loss_data)
    DATA_POINTE_ALL.append(loss_data2)
```
